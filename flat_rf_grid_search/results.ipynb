```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable
from sklearn.metrics import f1_score
```

```python
model_nums = []
weights = []
trues = []
preds = []
# Specify the directory path
directory_path = '/Users/adamboesky/Research/ay98/Weird_Galaxies/flat_rf_grid_search/results'

# Iterate through each file in the directory
model_num = 0
for filename in os.listdir(directory_path):
    file_path = os.path.join(directory_path, filename)
    with open(file_path, 'rb') as f:
        ws, (ts, ps) = pickle.load(f)
        model_nums.append(model_num)
        weights.append(ws)
        trues.append(np.concatenate(ts))
        preds.append(np.concatenate(ps))
        model_num += 1
```

```python
# Evaluate all the models by the f1_score
performance = {}
for m_num, ws, ts, ps in zip(model_nums, weights, trues, preds):
    performance[m_num] = f1_score(ts, ps, average='macro')

# Get the best model
best_mod_num = max(performance, key=performance.get)
print(f'Best model number is {best_mod_num}\n\tf1_score = {performance[best_mod_num]}\n\tweights = {weights[best_mod_num]}')
```
