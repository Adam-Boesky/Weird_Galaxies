```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

from astropy.io import fits
```

```python
test = fits.open('/Volumes/T7/ay_98_data/Zou_data/sed_catalog/es1.v1.fits')
```

```python
vals = test[1].data
```

```python
# ('detBIC1_agn', '>f8'), ('detBIC2_agn', '>f8'), ('detBIC1_bqgal', '>f8'), ('detBIC2_bqgal', '>f8')
plt.hist(np.log10(vals['detBIC1_agn']), alpha=0.5, label='detBIC1_agn')
plt.hist(np.log10(vals['detBIC2_agn']), alpha=0.5, label='detBIC2_agn')
plt.hist(np.log10(vals['detBIC1_bqgal'][vals['detBIC1_bqgal'] != 0]), alpha=0.5, label='detBIC1_bqgal')
plt.hist(np.log10(vals['detBIC2_bqgal']), alpha=0.5, label='detBIC2_bqgal')
plt.legend()
```

```python
plt.hist(np.log10(vals['Mstar_gal']), bins=50)
```

```python
# gal_typ = ['star' for v in vals['flag_star'] if v == 1]
gal_typ = []
for typ in zip(vals['flag_star'], vals['flag_Xrayagn'], vals['flag_IRagn_S05'], vals['flag_IRagn_L07'], vals['flag_IRagn_D12'], vals['flag_reliablesedagn'], vals['detBIC1_bqgal'], vals['detBIC2_bqgal'], vals['detBIC2_agn'], vals['detBIC1_agn']):
    if typ[0] == 1:
        gal_typ.append('star')
    elif typ[1] == 1:
        gal_typ.append('XrayAGN')
    elif typ[2] == 1:
        gal_typ.append('flag_IRagn_S05')
    elif typ[3] == 1:
        gal_typ.append('flag_IRagn_L07')
    elif typ[4] == 1:
        gal_typ.append('flag_IRagn_D12')
    elif typ[5] == 1:
        gal_typ.append('flag_reliablesedagn')
    elif typ[6] > 2 or typ[7] > 2:
        gal_typ.append('BQ')
    elif typ[8] > 2 or typ[9] > 2:
        gal_typ.append('AGN')
    else:
        gal_typ.append('other')


props_and_types = pd.DataFrame(np.vstack((vals['Mstar_best'].astype(float), vals['SFR_best'].astype(float), gal_typ)).T, columns=['Mstar_best', 'SFR_best', 'type'])
props_and_types = props_and_types.dropna(axis=1, how='any')
props_and_types = props_and_types[(props_and_types['Mstar_best'] != 'nan') & (props_and_types['SFR_best'] != 'nan')]
props_and_types['Mstar_best'] = np.log10(props_and_types['Mstar_best'].astype(float))
props_and_types['SFR_best'] = np.log10(props_and_types['SFR_best'].astype(float))
props_and_types['type'] = props_and_types['type'].astype(str)
props_and_types.head()
```

```python
props_and_types['type'].value_counts()
```

```python
sns.pairplot(props_and_types, x_vars=['Mstar_best', 'SFR_best'], y_vars=['Mstar_best', 'SFR_best'], hue='type')
```
